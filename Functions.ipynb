1.  What is the difference between a function and a method in Python?

In Python, both functions and methods are blocks of code that perform specific tasks, but they have key differences:

1. **Definition**:
   - **Function**: A function is a standalone block of code defined using the `def` keyword. It can be called independently.
   - **Method**: A method is a function that is associated with an object (an instance of a class) and is defined within a class. It operates on the data contained in that class.

2. **Calling**:
   - **Function**: Called by its name, e.g., `my_function()`.
   - **Method**: Called on an instance of a class, e.g., `my_object.my_method()`.

3. **Context**:
   - **Function**: Does not have access to the attributes of a class unless explicitly passed.
   - **Method**: Has access to the instance it is called on via `self`, allowing it to interact with the instance's attributes and other methods.

### Example:

```python
# Function
def greet(name):
    return f"Hello, {name}!"

# Method
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

# Usage
print(greet("John"))          # Calling a function
greeter = Greeter()
print(greeter.greet("Mike"))   # Calling a method
```

In summary, functions are independent, while methods are functions tied to class instances.

5. What are iterators in Python and how do they differ from iterables

In Python, **iterators** and **iterables** are closely related concepts, but they serve different purposes. Here's a breakdown of each:

### Iterable
- **Definition**: An iterable is any object in Python that can be looped over (iterated). This includes data types like lists, tuples, sets, dictionaries, and strings.
- **Behavior**: An iterable implements the `__iter__()` method, which returns an iterator object. It can also implement the `__getitem__()` method for indexed access.
- **Example**:
  ```python
  my_list = [1, 2, 3]
  for item in my_list:  # my_list is iterable
      print(item)
  ```

### Iterator
- **Definition**: An iterator is an object that represents a stream of data. It retrieves elements one at a time and maintains its state. An iterator implements two methods: `__iter__()` and `__next__()`.
- **Behavior**: The `__iter__()` method returns the iterator object itself, and the `__next__()` method returns the next value from the iterator. When there are no more items to return, `__next__()` raises a `StopIteration` exception.
- **Example**:
  ```python
  my_iter = iter(my_list)  # Creating an iterator from the iterable
  print(next(my_iter))  # Outputs: 1
  print(next(my_iter))  # Outputs: 2
  ```

### Key Differences
- **Definition**:
  - **Iterable**: Can be looped over (e.g., lists, strings).
  - **Iterator**: Provides the mechanism to iterate (e.g., the object created by `iter()`).
  
- **State**:
  - **Iterable**: Does not maintain state during iteration.
  - **Iterator**: Maintains its current position and state during iteration.

- **Usage**:
  - You can directly use iterables in loops. Iterators are usually obtained from iterables using the `iter()` function.

### Summary
In summary, an iterable is any object that can be iterated over, while an iterator is a specific object that keeps track of the current position during iteration. You can think of iterables as collections of data and iterators as the tools that allow you to access that data sequentially.

6. Explain the concept of generators in Python and how they are defined.

Generators in Python are a special type of iterator that allow you to create iterables in a more memory-efficient way. They provide a convenient way to define iterators using a function that yields values one at a time instead of returning all values at once.

Key Features of Generators:

Lazy Evaluation: Generators produce items one at a time and only when requested, which means they are more memory-efficient than storing all items in a list.

State Retention: When a generator function is called, it doesn’t start executing immediately. Instead, it returns a generator object that can be iterated over. The function's state is saved between calls, allowing it to resume where it left off.

Yield Statement: Instead of using return to return a value and terminate the function, a generator uses the yield statement. This allows the function to yield a value and pause its execution, resuming later when the next value is requested.

Defining a Generator:

You define a generator just like a regular function but use yield instead of return.

Example:

def count_up_to(max):
    count = 1
    while count <= max:
        yield count  # Yield the current count and pause execution
        count += 1   # Increment count

# Using the generator
counter = count_up_to(5)
for number in counter:
    print(number)  # Outputs: 1, 2, 3, 4, 5

How Generators Work
Creation: When you call count_up_to(5), it returns a generator object but does not start executing the function.

Iteration: Each time you iterate over the generator (e.g., using a loop), the function executes until it hits a yield statement. At this point, it yields the value and pauses execution.

Resumption: On the next iteration, execution resumes right after the last yield, continuing until the next yield or until the function ends.

**Summary**

Generators are a powerful feature in Python for creating iterators in a straightforward and efficient way.





7.  What are the advantages of using generators over regular functions?

Generators offer several advantages over regular functions, particularly when it comes to handling large datasets or streams of data. Here are the key benefits:

### 1. **Memory Efficiency**
- **Lazy Evaluation**: Generators produce values on the fly, which means they do not store all values in memory at once. This is particularly useful for large datasets, as it reduces memory consumption.
  
### 2. **Performance**
- **Reduced Overhead**: Since generators yield values one at a time, they can be faster for certain tasks compared to generating all values at once, especially when working with large data collections.

### 3. **Simpler Code**
- **Cleaner Syntax**: Generators can simplify code that requires iteration. Instead of managing state with a loop and a list, you can use `yield` to produce a sequence of values in a more concise manner.

### 4. **State Retention**
- **Persistent State**: Generators maintain their state between `yield` calls, allowing them to resume execution without needing to pass and return multiple values. This makes it easier to manage complex iterators.

### 5. **Infinite Sequences**
- **Handling Infinite Data**: Generators can represent infinite sequences (e.g., generating Fibonacci numbers) without consuming memory for all elements. Regular functions would struggle with this as they would need to return a complete collection.

### 6. **Pipelining**
- **Chaining Generators**: Generators can be easily chained together to create pipelines of operations (e.g., filtering and transforming data). Each generator can pass its output to the next generator in a memory-efficient way.

### Example of Chaining Generators:
```python
def generate_numbers(n):
    for i in range(n):
        yield i

def square_numbers(nums):
    for num in nums:
        yield num ** 2

nums = generate_numbers(5)
squared = square_numbers(nums)

for number in squared:
    print(number)  # Outputs: 0, 1, 4, 9, 16
```

### Summary
In summary, generators provide significant advantages in terms of memory efficiency, performance, code simplicity, state management, and the ability to work with infinite sequences. They are particularly useful in scenarios where data processing is iterative or where large amounts of data need to be handled without overwhelming system resources.

8. What is a lambda function in Python and when is it typically used?

A **lambda function** in Python is an anonymous function defined using the `lambda` keyword. Unlike regular functions defined with `def`, lambda functions are typically small and can take any number of arguments but can only have a single expression. The result of this expression is automatically returned.

### Syntax
The syntax of a lambda function is:
```python
lambda arguments: expression
```

### Example
```python
# A simple lambda function that adds two numbers
add = lambda x, y: x + y
print(add(3, 5))  # Output: 8
```

### When to Use Lambda Functions
Lambda functions are typically used in the following scenarios:

1. **Short-lived Functions**: When you need a simple function for a short period and don't want to formally define it with a name using `def`.

2. **Higher-Order Functions**: They are often used as arguments to higher-order functions (functions that take other functions as arguments), such as:
   - `map()`: Applies a function to all items in an iterable.
   - `filter()`: Filters items in an iterable based on a condition.
   - `sorted()`: Custom sorting with a key function.

   **Example with `map()`**:
   ```python
   numbers = [1, 2, 3, 4]
   squares = list(map(lambda x: x ** 2, numbers))
   print(squares)  # Output: [1, 4, 9, 16]
   ```

3. **Inline Function Definition**: When you need to define a function that will be used once or in a specific context, such as in a list comprehension or an event handler.

4. **Simplifying Code**: They can help make code more concise, especially for small operations where defining a full function would be unnecessarily verbose.

### Limitations
- **Single Expression**: Lambda functions can only contain a single expression, making them less versatile than regular functions.
- **Readability**: While they can make code shorter, overusing them can lead to less readable code, especially if the logic is complex.

### Summary
In summary, lambda functions are a convenient way to create simple, anonymous functions for use in functional programming contexts or when defining short-lived functions, making your code more concise and flexible.

9.  Explain the purpose and usage of the `map()` function in Python

The `map()` function in Python is a built-in function that allows you to apply a specified function to all items in an iterable (such as a list, tuple, or string). The primary purpose of `map()` is to transform data in a concise and efficient way.

### Purpose
- **Transformation**: `map()` is used to apply a transformation function to each item in the iterable, producing a new iterable with the results.
- **Avoiding Loops**: It provides a more functional programming style, reducing the need for explicit loops and making code more concise.

### Syntax
```python
map(function, iterable, ...)
```
- **function**: A function that takes one or more arguments (depending on the number of iterables passed) and returns a value.
- **iterable**: One or more iterables (lists, tuples, etc.) to which the function will be applied.

### Returns
`map()` returns a map object (an iterator) which can be converted into a list, tuple, or another iterable.

### Example Usage

1. **Single Iterable**:
   ```python
   numbers = [1, 2, 3, 4]
   squares = map(lambda x: x ** 2, numbers)  # Applying a lambda function to square each number
   print(list(squares))  # Output: [1, 4, 9, 16]
   ```

2. **Multiple Iterables**:
   You can also pass multiple iterables to `map()`, in which case the function should accept as many arguments as there are iterables.
   ```python
   list1 = [1, 2, 3]
   list2 = [4, 5, 6]
   sums = map(lambda x, y: x + y, list1, list2)  # Summing elements from both lists
   print(list(sums))  # Output: [5, 7, 9]
   ```

3. **Using Named Functions**:
   Instead of a lambda function, you can use a named function.
   ```python
   def square(x):
       return x ** 2

   numbers = [1, 2, 3, 4]
   squares = map(square, numbers)
   print(list(squares))  # Output: [1, 4, 9, 16]
   ```

### Advantages
- **Conciseness**: It allows you to apply a function to each item in an iterable in a single line of code.
- **Efficiency**: It can be more efficient than using a for-loop, especially with large datasets, since it generates results on-the-fly (lazy evaluation).

### Summary
In summary, the `map()` function is a powerful tool for transforming data in iterables. It promotes a functional programming style and can help make your code cleaner and more readable when dealing with bulk transformations.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

The `map()`, `reduce()`, and `filter()` functions in Python are all higher-order functions that allow you to work with iterables in a functional programming style. However, they serve different purposes:

### 1. `map()`
- **Purpose**: Transforms each item in an iterable by applying a specified function.
- **Returns**: A map object (iterator) containing the results of applying the function to each item.
- **Usage**: When you want to apply a function to all elements in an iterable.

**Example**:
```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```

### 2. `filter()`
- **Purpose**: Filters items in an iterable based on a specified function that returns `True` or `False`.
- **Returns**: A filter object (iterator) containing only the items for which the function returned `True`.
- **Usage**: When you want to select a subset of elements from an iterable based on a condition.

**Example**:
```python
numbers = [1, 2, 3, 4]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4]
```

### 3. `reduce()`
- **Purpose**: Applies a specified function cumulatively to the items of an iterable, reducing it to a single value.
- **Returns**: A single value that results from the cumulative application of the function.
- **Usage**: When you want to combine all elements of an iterable into a single value (e.g., summing numbers).

**Example**:
```python
from functools import reduce

numbers = [1, 2, 3, 4]
sum_result = reduce(lambda x, y: x + y, numbers)
print(sum_result)  # Output: 10
```

### Summary of Differences
- **`map()`**: Transforms each element of an iterable using a function; returns an iterable of the same length.
- **`filter()`**: Filters elements based on a condition; returns an iterable of elements that meet the condition.
- **`reduce()`**: Reduces an iterable to a single cumulative value; returns a single value based on the function applied to the elements.

These functions are powerful tools for data manipulation and can lead to more concise and expressive code in Python.

4. Write a Python function that checks if a given number is prime or not from 1 to 200

````python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Example usage:
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is prime")
````

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

````python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next_value = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            fib_value = self.current
            self.current, self.next_value = self.next_value, self.current + self.next_value
            self.count += 1
            return fib_value
        else:
            raise StopIteration

# Example usage:
fib = FibonacciIterator(10)  # Change the number to generate more or fewer terms
for number in fib:
    print(number)
````

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

````python
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Using map to convert the temperatures
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Example usage
print(fahrenheit_temps)  # Output: [32.0, 68.0, 98.6, 212.0]
````

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
# Function to filter out vowels
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, s))

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)

print(result)  # Output: Hll, Wrld!
````
