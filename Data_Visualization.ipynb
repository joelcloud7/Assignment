```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a synthetic dataset with categories and corresponding numerical values
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = {
    'Category A': np.random.normal(50, 10, 100),  # Mean = 50, Std = 10, 100 samples
    'Category B': np.random.normal(60, 15, 100),  # Mean = 60, Std = 15, 100 samples
    'Category C': np.random.normal(55, 5, 100),   # Mean = 55, Std = 5, 100 samples
    'Category D': np.random.normal(65, 20, 100),  # Mean = 65, Std = 20, 100 samples
}

# Convert the dataset into a format that seaborn can use (long-form)
import pandas as pd
df = pd.DataFrame({
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate([values[cat] for cat in categories])
})

# Create a boxplot to visualize the distribution of the values across categories
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')

# Add title and labels
plt.title('Distribution of Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')

# Show the plot
plt.show()
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Create a synthetic dataset with correlated features
n_samples = 1000

# Generating random data
x1 = np.random.normal(0, 1, n_samples)
x2 = 0.7 * x1 + np.random.normal(0, 0.5, n_samples)  # x2 is correlated with x1
x3 = 0.3 * x1 + 0.5 * x2 + np.random.normal(0, 0.2, n_samples)  # x3 is correlated with x1 and x2
x4 = np.random.normal(0, 1, n_samples)  # x4 is independent

# Creating a DataFrame with these variables
data = pd.DataFrame({
    'x1': x1,
    'x2': x2,
    'x3': x3,
    'x4': x4
})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Plotting the heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.




```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Use output_notebook() to display Bokeh plots in Colab
output_notebook()

# Generate random data
np.random.seed(42)

# Number of points
n = 100

# Random x and y values
x = np.random.random(n) * 10  # Random x-values between 0 and 10
y = np.random.random(n) * 10  # Random y-values between 0 and 10

# Random sizes (between 5 and 50 for marker size)
sizes = np.random.randint(5, 50, size=n)

# Random colors (choose from a list of colors)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)

# Create a DataFrame for the data
data = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

# Create ColumnDataSource
source = ColumnDataSource(data)

# Create a figure
p = figure(title="Bokeh Scatter Plot with Legend", x_axis_label='x', y_axis_label='y', width=800, height=400)

# Add scatter plot with different sizes and colors from ColumnDataSource
p.scatter(x='x', y='y', size='sizes', color='colors', legend_field='colors', source=source, fill_alpha=0.6)

# Show the plot inline in Colab
show(p)
```

```python
# Step 1: Import necessary libraries
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Viridis256
from bokeh.models import ColumnDataSource, ColorBar
from bokeh.transform import linear_cmap
from bokeh.layouts import column
from bokeh.io import push_notebook

# Step 2: Use output_notebook() to display the plot inline in Colab
output_notebook()

# Step 3: Generate the data
data_heatmap = np.random.rand(10, 10)  # Random 10x10 data matrix
x = np.linspace(0, 1, 10)  # x-axis values
y = np.linspace(0, 1, 10)  # y-axis values
xx, yy = np.meshgrid(x, y)  # Create a grid

# Step 4: Flatten the meshgrid for Bokeh ColumnDataSource
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Step 5: Create ColumnDataSource for Bokeh
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, z=z_flat))

# Step 6: Define a color map using linear_cmap to map z values to colors
color_map = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

# Step 7: Create a figure for the heatmap
p = figure(title="Heatmap Example",
           x_axis_label='X',
           y_axis_label='Y',
           width=800, height=800)

# Step 8: Add rectangles (heatmap) to the plot
p.rect(x='x', y='y', width=1/10, height=1/10, source=source,
       color=color_map, line_color=None)

# Step 9: Add color bar to the plot
color_bar = ColorBar(color_mapper=color_map['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Step 10: Show the plot
show(p, notebook_handle=True)
```
